```python
# import Opencv
import cv2
import numpy as np  
from matplotlib import pyplot as plt
# read a image using imread
img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Computer_Vision_dr/rose.jpg', 0)
  
# creating a Histograms Equalization of an image using cv2.equalizeHist()
equ = cv2.equalizeHist(img)
  
# stacking images side-by-side
res = np.hstack((img, equ))
  
# show image input vs output

plt.figure(figsize=(10,10))

# cv2_imshow(img)
# cv2_imshow(res)

plt.imshow(img)
plt.imshow(res)
plt.axis('off')
plt.show()
# cv2_imshow('/content/drive/MyDrive/Colab Notebooks/Computer_Vision_dr/free_crim.jpg', res)
  
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import traitlets
img_fl =cv2.imread('/content/drive/MyDrive/Colab Notebooks/Computer_Vision_dr/rose3.jpg')
img_fl = cv2.cvtColor(img_fl, cv2.COLOR_BGR2RGB)
R , G, B = cv2.split(img_fl)

output1_R = cv2.equalizeHist(R)
output1_G = cv2.equalizeHist(G)
output1_B = cv2.equalizeHist(B)
output1 = cv2.merge((output1_R,output1_G, output1_B))

clahe = cv2.createCLAHE(clipLimit = 5.0 , tileGridSize=(16,16))

output2_R = cv2.equalizeHist(R)
output2_G = cv2.equalizeHist(G)
output2_B = cv2.equalizeHist(B)

output2 = cv2.merge((output2_R,output2_G, output2_B))

output = [img_fl, output1,output2]
titles = ['original' ,'Adjusted Histogram','CLAHE']

f = plt.figure()
f.set_figwidth(15)
f.set_figheight(15)

for i in range(3):

  plt.subplot(1,3,i+1)
  plt.imshow(output[i])
  plt.title(titles[i])
  plt.xticks([])
  plt.yticks([])
plt.show()

#histogram Equalization after changing the image to grayscale
img_chng = cv2.cvtColor(img_fl, cv2.COLOR_BGR2GRAY)
plt.hist(img_fl.flat , bins = 100 , range=(0,255))
equ = cv2.equalizeHist(img_chng)

plt.hist(equ.flat , bins = 100 , range=(0,255))
plt.xlabel("Intensity")
plt.ylabel("Frequency")
```

Histogram equalization and clahe of low contrast image


```python
img2 = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Computer_Vision_dr/rose_low_cont.jpg')
lab_img = cv2.cvtColor(img2 , cv2.COLOR_BGR2LAB)
cv2_imshow(img2)
l , a , b = cv2.split(lab_img)
# print("l",l, '\n',"a",a,'\n',"b",b)

plt.hist(l.flat , bins = 100 , range=(0,255))
equ = cv2.equalizeHist(l)

plt.hist(equ.flat , bins = 100 , range=(0,255))
plt.xlabel("Intensity")
plt.ylabel("Frequency")
```

```python
plt.imshow(equ,cmap ='gray')
```

```python
updated_lab_img1 = cv2.merge((equ,a,b))
hist_eq_img = cv2.cvtColor(updated_lab_img1, cv2.COLOR_LAB2BGR)

#Apply clahe to L channel
clahe = cv2.createCLAHE(clipLimit = 3.0 , tileGridSize =(8,8))
clahe_img = clahe.apply(l)

updated_lab_img2 = cv2.merge((clahe_img,a,b))
#Converting lab image back to color image
CLAHE_img = cv2.cvtColor(updated_lab_img2 , cv2.COLOR_LAB2BGR)

cv2_imshow(img2)
# "Equalized Image" 
cv2_imshow( hist_eq_img)
# "Clahe Image" 
cv2_imshow( CLAHE_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
