```python
import cv2 
import numpy as np 
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

# Translation

```python
 # Read image from disk 
img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Computer_Vision_dr/rose3.jpg',0) 

rows, cols = img.shape[:2] 
cv2_imshow(img)

'''
# The shape of an image is accessed by img.shape. 
 It returns a tuple of the number of rows, columns, and channels (if the image is color)
# Creating translation matrix for shifting the image in the location(100, 50). 100 -> coloumn ; 50 ->row
# If the shift is (x, y) then matrix would be
# M = [1 0 x]
#     [0 1 y]   #if we change the value image will come as zoom or floating
'''

M = np.float32([[1, 0, 100], [0, 1, 150]])        #original image just change the axis
M1 = np.float32([[1, 0, -100], [0, 1, 150]])      #change of axis from negative
# M1 = np.float32([[0, 1, -100], [1, 0, 150]])    #kind of rotated image 
# M1 = np.float32([[5, 0, 100], [0, 1, 150]])   #pretend zoomed image


```

```python
# warpAffine does appropriate shifting given the
# translation matrix.
res = cv2.warpAffine(img, M, (cols, rows))
res1 = cv2.warpAffine(img, M1, (cols, rows))


plt.subplot(131),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(res, cmap = 'gray')
plt.title('Output image'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(res1, cmap = 'gray')
plt.title('Output image2'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
  # Write image back to disk. 
cv2.imwrite('result.jpg', res) 
```

#Rotation 

```python
# Read image from the disk. 
img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Computer_Vision_dr/rose3.jpg',0) 

print(img.shape)
# Shape of image in terms of pixels. 
(rows, cols) = img.shape[:2] 
print(rows,cols)
```

```python
# getRotationMatrix2D creates a matrix needed for transformation. 
#Rotate 45 degree with respect to center
# cv2.getRotationMatrix2D(center, angle, scale)
M = cv2.getRotationMatrix2D((cols / 2, rows / 2), 45,1) 
M2 = cv2.getRotationMatrix2D((cols / 2, rows / 2), 45,2)    #positive rotation
M3 = cv2.getRotationMatrix2D((cols / 2, rows / 2), 45,-2)   #negative rotation
M3 = cv2.getRotationMatrix2D((cols / 5, rows / 5), 45,1)    #dividing the coloumn and rows in that part and rotating


#Wrap affine - >
res = cv2.warpAffine(img, M, (cols, rows)) 
res2 = cv2.warpAffine(img, M2, (cols, rows)) 
res3 = cv2.warpAffine(img, M3, (cols, rows)) 

# cmap2 = plt.cm.get_cmap(res)
# print(cmap2)
plt.subplot(141),plt.imshow(img, cmap = 'jet')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(142),plt.imshow(res, cmap = 'gray')
plt.title('Output1'), plt.xticks([]), plt.yticks([])
plt.subplot(143),plt.imshow(res2, cmap = 'gray')
plt.title('Output2'), plt.xticks([]), plt.yticks([])
plt.subplot(144),plt.imshow(res3, cmap = 'gray')
plt.title('Output3'), plt.xticks([]), plt.yticks([])
plt.show()
```

# **#Geometric transformation-scalling**


```python
# Read image from disk. 
img1 = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Computer_Vision_dr/rose3.jpg',0)
rows, cols = img1.shape

# Get number of pixel horizontally and vertically. 
(height, width) = img1.shape[:2] 
print(height,width)
```

```python
# Specify the size of image along with interpolation methods. 
# cv2.INTER_AREA is used for shrinking, whereas cv2.INTER_CUBIC is used for zooming. 
res = cv2.resize(img1, (int(width / 4), int(height / 2)), interpolation = cv2.INTER_CUBIC) 
crows, ccols = res.shape
plt.subplot(121),plt.imshow(img1, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(res, cmap = 'gray')
plt.title('Output image'), plt.xticks([]), plt.yticks([])
plt.show()
cv2_imshow(img1)
cv2_imshow(res)
```
