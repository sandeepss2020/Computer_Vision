```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread("/content/drive/MyDrive/Colab Notebooks/Computer_Vision_dr/criminal2.jpg", cv2.IMREAD_COLOR)
```

```python
cv2_imshow(img)
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_img)
```

```python
cv2.imwrite("/content/drive/MyDrive/Colab Notebooks/Computer_Vision_dr/criminal2_updated.jpg", gray_img)
print("Image created!")
```
