```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Computer_Vision_dr/coin1.jpg')
images = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(images, cv2.COLOR_RGB2GRAY)
# cv2_imshow(image)
```

```python
plt.figure(figsize=(25, 5), constrained_layout=False)
plt.subplot(131), plt.imshow(image), plt.title("Original image")
plt.axis('off')
plt.subplot(132), plt.imshow(images), plt.title("RGB")
plt.axis('off')
plt.subplot(133), plt.imshow(gray), plt.title("GRAY IMAGE")
plt.axis('off')
```

```python
image2 = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Computer_Vision_dr/coin2.jpg')
images2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
gray2 = cv2.cvtColor(images2, cv2.COLOR_RGB2GRAY)
```

```python
plt.figure(figsize=(25, 5), constrained_layout=False)
plt.subplot(131), plt.imshow(image2), plt.title("Original image")
plt.axis('off')
plt.subplot(132), plt.imshow(images2), plt.title("RGB")
plt.axis('off')
plt.subplot(133), plt.imshow(gray2), plt.title("GRAY IMAGE")
plt.axis('off')
```

```python
# create a binary thresholded image
_, binary = cv2.threshold(gray, 225, 255, cv2.THRESH_BINARY_INV)   #Detected edges may supply as input
_, binary2 = cv2.threshold(gray2, 225, 255, cv2.THRESH_BINARY_INV)   #Detected edges may supply as input

#<225 is set to 0 and >=225 is set to 255 show it
plt.imshow(binary, cmap="gray")
plt.show()

plt.imshow(binary2, cmap="gray")
plt.show()

# plt.figure(figsize=(25, 5), constrained_layout=False)
# plt.subplot(121), plt.imshow(binary), plt.title("Binary")
# plt.axis('off')
# plt.subplot(122), plt.imshow(binary2), plt.title("Binary2")
# plt.axis('off')
```

```python
# find the contours from the thresholded image
contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

contours2, hierarchy2 = cv2.findContours(binary2, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# RETR_EXTERNAL, contour retrival mode
#cv2.RETR_LIST: For extracting all contours without establishing any relationships
#CHAIN_APPROX_NONE -contour approxmation method-all boundary points are stored suppose 
#if we want to find contour of stright line then two points are sufficient so we give simple
# none will store all boundary points
image = cv2.drawContours(image, contours, -1, (0, 255, 0), 2)#-1 display all contours
image2 = cv2.drawContours(image2, contours2, -1, (0, 255, 0), 2)#-1 display all contours

# show the image with the drawn contours
plt.imshow(image)
plt.show()

plt.imshow(image2)
plt.show()
```
